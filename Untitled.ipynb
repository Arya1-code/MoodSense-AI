```python
from pathlib import Path
import pandas as pd
dataset_path = Path("datasets/raw/fer2013/train")

counts = {}

for emotion in dataset_path.iterdir():
    if emotion.is_dir():
        counts[emotion.name] = len(list(emotion.glob("*")))

df = pd.DataFrame(
    counts.items(),
    columns=["Emotion", "Count"]
)

df
```

```python
import matplotlib.pyplot as plt

df.sort_values("Count").plot(
    x="Emotion",
    y="Count",
    kind="bar"
)

plt.title("FER2013 Class Distribution")
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path

train_path = Path("datasets/raw/fer2013/train")

fig, axes = plt.subplots(1, 7, figsize=(18, 4))

for ax, emotion in zip(axes, sorted(train_path.iterdir())):
    if emotion.is_dir():
        img_path = next(emotion.glob("*"))
        img = Image.open(img_path)

        ax.imshow(img, cmap="gray")
        ax.set_title(emotion.name)
        ax.axis("off")

plt.tight_layout()
plt.show()
```
